```python
import string
import nltk 
import pandas as pd
import re
from collections import defaultdict, Counter
nltk.download('brown')
nltk.download('punkt')
from nltk.corpus import brown
from nltk import word_tokenize
```

```python
# Step 1 : retrieve data
tokens = brown.words()[:200000]
print(tokens[:20])
print(len(tokens))
```

```python
# Step 2 : Data Preprocessing
def preprocess_tokens(text):
    clean_tokens = []
    for word in text:
        text_1 = word.lower()
        text_1 = re.sub(f"[{re.escape(string.punctuation)}]","",text_1)
        if all(char in string.punctuation for char in word) or len(text_1) == 0:
            continue
        clean_tokens.append(text_1)
    return clean_tokens

clean_tokens = preprocess_tokens(tokens)

clean_tokens = set(clean_tokens)
# print(tokens[:20])
print(len(clean_tokens))

```

```python
# Step 3 : Build the N-gram Models 
def n_gram(tokens,n):
    ngram = []
    for i in range(len(tokens)-n+1):
        ngram.append(tuple(tokens[i:i+n]))
        
    ngram_counts={}
    for i in range(len(tokens)-n+1):
        pair = tuple(tokens[i:i+n])
        if pair not in ngram_counts:
            ngram_counts[pair] = 1
        else:
            ngram_counts[pair] += 1
    prefix_counts = {}
    for ngram in ngram_counts:
        prefix = ngram[:-1]
        if prefix not in prefix_counts:
            prefix_counts[prefix] = ngram_counts[ngram]
        else:
            prefix_counts[prefix] += ngram_counts[ngram]            
            
        
```
